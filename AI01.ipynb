# 畫函數圖形
\begin{equation}
x = \sin(p t), \quad y = \cos(q t)
\end{equation}
- 這個圖形是一條 **李薩如曲線 (Lissajous Curve)**，描述兩個互相垂直的簡諧運動之間的相互關係。
- 線條顏色從金黃色漸變至棕色，搭配圖形看起來變成一塊菠蘿麵包。
- 可以透過調整參數來改變函數圖形的顯示：
    1. **頻率比值 $p/q$**：影響曲線的形狀與對稱性
    2. **點的數量**：決定曲線的平滑程度

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# 定義李薩如曲線的參數
p, q = 6, 7
t = np.linspace(0, 2 * np.pi, 1000)
x = np.sin(p * t)
y = np.cos(q * t)

# 建立金黃色到棕色的漸變色
colors = ["gold", "saddlebrown"]
cmap = LinearSegmentedColormap.from_list("golden_brown", colors, N=len(t))

# 繪製李薩如曲線
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(len(t) - 1):
    ax.plot(x[i:i+2], y[i:i+2], color=cmap(i / len(t)), linewidth=2)

# 設定圖形外觀
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
ax.set_facecolor("white")

# 顯示圖形
plt.show()
```
