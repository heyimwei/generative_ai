# 載入必要套件

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 畫函數圖形
\begin{equation}
x = \sin(p t), \quad y = \cos(q t)
\end{equation}
- 這個圖形是一條 **李薩如曲線 (Lissajous Curve)**，描述兩個互相垂直的簡諧運動之間的相互關係。
- 線條顏色從金黃色漸變至棕色，搭配圖形看起來變成一塊菠蘿麵包。
- 為了使視覺效果不受干擾，我沒有繪製格線與座標軸。
- 可以透過調整參數來改變函數圖形的顯示：
    1. **頻率比值 $p/q$**：影響曲線的形狀與對稱性
    2. **點的數量**：決定曲線的平滑程度

```python
from matplotlib.colors import LinearSegmentedColormap  # 用於自定義漸變色

# 定義李薩如曲線的參數
p, q = 6, 7  # 決定曲線的形狀
t = np.linspace(0, 2 * np.pi, 2000)
x = np.sin(p * t)  # 算x座標
y = np.cos(q * t)  # 算y座標

# 設定顏色
colors = ["gold", "saddlebrown"]  # 定義漸變色範圍從金黃色到棕色
cmap = LinearSegmentedColormap.from_list("golden_brown", colors, N=len(t))  # 建立漸變色

# 繪製李薩如曲線
fig, ax = plt.subplots(figsize=(6, 6))  # 設定大小，建立畫布
for i in range(len(t) - 1):  # 逐點繪製曲線
    ax.plot(x[i:i+2], y[i:i+2], color=cmap(i / len(t)), linewidth=2)  # 設定漸變色

# 顯示圖形
plt.show()  # 顯示繪製的李薩如曲線
```
